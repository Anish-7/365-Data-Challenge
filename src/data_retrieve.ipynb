```python
import mysql.connector

db = mysql.connector.connect(host='localhost',user='root',password='mynewpassword',database='365_database',auth_plugin='mysql_native_password')

mycurzor=db.cursor()
```

```python
import pandas as pd
query ="""
SELECT  
student_id

FROM 
365_student_info

GROUP BY 
student_id
"""
ids_list =pd.read_sql(query,db)
id=ids_list.student_id.unique()
ids_list
```

```python
import pandas as pd

query ="""
SELECT  
student_id ,AVG(minutes_watched) as avg_time

FROM 
365_student_learning

GROUP BY 
student_id
"""
watch_avg =pd.read_sql(query,db)
watch_avg_list = watch_avg.student_id.unique()
watch_avg_list

wa_re=[]
for i in id:
    if i not in watch_avg_list:
        wa_re.append([i,0])    


wa_re = pd.DataFrame(wa_re,columns=['student_id','avg_time'])    
features=watch_avg.append(wa_re)

features
```

```python
import pandas as pd

query ="""
SELECT  
DISTINCT student_id
FROM 
365_student_quizzes
"""

quiz_taken =pd.read_sql(query,db)
quiz_taken_id=quiz_taken.student_id.unique()
qt_re=[]
for i in id:
    if i in quiz_taken_id:
        qt_re.append([i,1])
    else:
        qt_re.append([i,0])

qt_re = pd.DataFrame(qt_re,columns=['student_id','taken_quiz'])    
features=pd.merge(features,qt_re,on='student_id',how='outer')

features
```

```python
query ="""
SELECT  
DISTINCT student_id
FROM 
365_student_exams
"""

exam_taken =pd.read_sql(query,db)
exam_taken_id=exam_taken.student_id.unique()

et_re=[]

for i in id:
    if i in exam_taken_id:
        et_re.append([i,1])
    else:
        et_re.append([i,0])

et_re = pd.DataFrame(et_re,columns=['student_id','exam_quiz'])    
features=pd.merge(features,et_re,on='student_id',how='outer')
features
```

```python
query ="""
SELECT  
DISTINCT student_id
FROM 
365_student_hub_questions
"""
hb_re=[]
hub_taken =pd.read_sql(query,db)
hub_taken_id=hub_taken.student_id.unique()

for i in id:
    if i in hub_taken_id :
        hb_re.append([i,1])
    else:
        hb_re.append([i,0])


hb_re = pd.DataFrame(hb_re,columns=['student_id','hub_quiz'])    
features=pd.merge(features,hb_re,on='student_id',how='outer')
features
```

```python
# Retriving the TARGET VARIABLE

query = ''' 

SELECT student_id 
FROM 365_student_purchases
'''

purchase_df = pd.read_sql(query ,db)
purchased  = purchase_df.student_id.unique()

p_res =[]

for i in id:
    if i in purchased:
        p_res.append([i,1])
    else :
        p_res.append([i,0])

p_res = pd.DataFrame(p_res,columns=['student_id','purchased'])
features = pd.merge(features,p_res,on='student_id',how='outer')
```
